### Exploratory data Analysis

```python
import networkx as nx
graph = nx.read_edgelist("D:\Downloads\SMA\com-lj.ungraph modified.txt")
```

# Calculating basic statistics

```python
# Calculating basic statistics
num_nodes = graph.number_of_nodes()
num_edges = graph.number_of_edges()
graph_density = nx.density(graph)

print("Number of nodes:", num_nodes)
print("Number of edges:", num_edges)
print("Graph density:", graph_density)
```

# Finding Centrality

```python
degree_centrality = nx.degree_centrality(graph)
betweenness_centrality = nx.betweenness_centrality(graph)
closeness_centrality = nx.closeness_centrality(graph)
eigenvector_centrality = nx.eigenvector_centrality(graph)


sorted_degree_centrality = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)
print("Top 5 nodes by degree centrality:")
for node, centrality in sorted_degree_centrality[:5]:
    print(f"Node {node}: Degree Centrality {centrality}")

sorted_betweenness_centrality = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)
print("Top 5 nodes by betweenness centrality:")
for node, centrality in sorted_betweenness_centrality[:5]:
    print(f"Node {node}: Betweenness Centrality {centrality}")
    

# Sort and print the top 5 nodes by closeness centrality
sorted_closeness_centrality = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)
print("Top 5 nodes by Closeness centrality:")
for node, centrality in sorted_closeness_centrality[:5]:
    print(f"Node {node}: Closeness Centrality {centrality}")
    

# Sort and print the top 5 nodes by eigenvector centrality
sorted_eigenvector_centrality = sorted(eigenvector_centrality.items(), key=lambda x: x[1], reverse=True)
print("Top 5 nodes by Eigenvector Centrality:")
for node, centrality in sorted_eigenvector_centrality[:5]:
    print(f"Node {node}: Eigenvector Centrality {centrality}")
```

# Network Metrics and Analysis

```python
#The average shortest path length characterizes the average distance between pairs of nodes in the network. It provides insights into how quickly information or influence can spread across the network.
if nx.is_connected(graph):
    avg_path_length = nx.average_shortest_path_length(graph)
    print("Average Path Length:", avg_path_length)
else:
    print("Graph is not connected") 
```

```python
#The clustering coefficient measures the degree to which nodes in the network tend to cluster together. It indicates the presence of tightly-knit communities or cliques within the network.
clustering_coefficient = nx.average_clustering(graph)
print("Clustering Coefficient:", clustering_coefficient)
```

```python
#Assortativity measures the tendency of nodes to connect to similar nodes in terms of certain attributes, such as degree or other node characteristics. It can be computed using the nx.degree_assortativity_coefficient function.
assortativity = nx.degree_assortativity_coefficient(graph)
print("Assortativity:", assortativity)
```

```python
if nx.is_connected(graph):
    diameter = nx.diameter(graph)
    print("Diameter of the Graph:", diameter)
else:
    print("The graph is not connected.")
```

# Connected Components

```python
connected_components = list(nx.connected_components(graph))
print(f'Total connected components are {len(connected_components)}')
for i, component in enumerate(connected_components):
    subgraph = graph.subgraph(component)
    diameter = nx.diameter(subgraph)
    print(f"Diameter of Component {i+1}: {diameter}")
```

## Preprocessing 2nd dataset 

```python
# Step 1: Read the first dataset (edges) to extract all nodes present in the graph
nodes_in_graph = set()
with open("D:\Downloads\SMA\com-lj.ungraph modified.txt", 'r') as edges_file:
    for line in edges_file:
        node1, node2 = map(str, line.strip().split('\t'))
        nodes_in_graph.add(node1)
        nodes_in_graph.add(node2)
# print('Nodes', nodes_in_graph)
# Step 2: Read the second dataset (communities) and filter out nodes
filtered_communities = []
with open("D:\Downloads\SMA\community.txt", 'r') as communities_file:
    for line in communities_file:
        community = line.strip().split()

        filtered_community = [node for node in community if node in nodes_in_graph]
#         print(filtered_community)
        if len(filtered_community)>2:  # Check if the filtered community is not empty
            filtered_communities.append(filtered_community)
# print(filtered_communities)

# Step 3: Save the filtered communities to a new file
with open('filtered_communities.txt', 'w') as filtered_file:
    for community in filtered_communities:
        filtered_file.write(' '.join(community) + '\n')

print("Filtered communities have been saved to 'filtered_communities.txt'")
```

## Analysis on 2nd Dataset

```python
import networkx as nx
import itertools
import statistics

# Load your dataset from a text file (replace 'your_dataset.txt' with your file path)
with open(r"D:\Downloads\SMA\filtered_communities.txt", 'r') as file:
    lines = file.readlines()

# Create an empty graph
G = nx.Graph()

# Create a list to store community information
communities = []

# Parse your data and add nodes to communities
for line in lines:
    nodes = line.strip().split()  # Assuming nodes are separated by spaces

    community = set(nodes)
    communities.append(community)
    # Add edges within the community for overlap analysis
    G.add_edges_from(itertools.combinations(community, 2))

# Calculate the total number of communities
num_communities = len(communities)

# Calculate min and max community sizes
community_sizes = [len(community) for community in communities]

min_community_size = min(community_sizes)
max_community_size = max(community_sizes)

# Calculate community overlap
overlap = len(G.edges()) - len(G) + num_communities

# Calculate centrality metrics for the entire graph or individual communities
# For example, degree centrality
centrality = nx.degree_centrality(G)

# Calculate clustering coefficient for the entire graph or individual nodes
# For example, clustering coefficient for each node
clustering_coefficients = nx.clustering(G)

# Calculate other relevant statistics as needed

# Print the results
print(f"Total Number of Communities: {num_communities}")
print(f"Minimum Community Size: {min_community_size}")
print(f"Maximum Community Size: {max_community_size}")
print(f"Community Overlap: {overlap}")
print(f"Centrality Metrics: {centrality}")
print(f"Clustering Coefficients: {clustering_coefficients}")
```
