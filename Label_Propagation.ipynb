```python
import networkx as nx

top_1000_nodes_file_path = 'top_1000_nodes.txt'
edge_pairs_file_path = 'Filtered_Edge_Pairs.txt'

with open(top_1000_nodes_file_path, 'r') as file:
    top_1000_nodes = [line.strip() for line in file]

edges = []
with open(edge_pairs_file_path, 'r') as file:
    for line in file:
        node1, node2 = line.strip().split()  # Assuming space-separated values
        edges.append((node1, node2))

G = nx.Graph()
G.add_nodes_from(top_1000_nodes)
G.add_edges_from(edges)
```

```python
G
```

```python
from networkx.algorithms import community

communities = community.label_propagation_communities(G)
communities = list(communities)  

with open('lpa_communities.txt', 'w') as file:
    for community in communities:
        community_str = '\t'.join(community)
        file.write(community_str + '\n')
```
